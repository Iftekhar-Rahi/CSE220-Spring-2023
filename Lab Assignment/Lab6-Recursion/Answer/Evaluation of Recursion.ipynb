Write a python code that will do the following:
1. Write a LinkedList class where the constructor will take an array of integers where all the
values are sorted in ascending order and create a non-dummy headed singly linked list.
2. You need to create a Node class too for the singly linked list
3. Write a recursive function named ‘remove_duplicates’ that will take the head of the
linked list as a parameter and remove all the duplicate values recursively

Example - 1:
Sample input: 10 → 20 → 20 → 30 → None
Sample output: 10 → 20 → 30 → None

Example - 2:
Sample input: 10 → 20 → 20 → 20 → 30 → None
Sample output: 10 → 20 → 30 → None

```python
arr=[10,20,30,40]
class Node:
  def __init__(self,value):
    self.value=value
    self.next=None
class LinkedList:
  def __init__(self,arr):
    self.head=None
    self.tail=None
    self.head=self.create_LL(arr)

  def create_LL(self,arr,i=0):
    if i==len(arr):
      return
    if i==0:
      # print(i)
      self.head=Node(arr[i])
      self.tail=self.head
    else:
      # print(i)
      temp=Node(arr[i])
      self.tail.next=temp
      self.tail=self.tail.next
    self.create_LL(arr,i+1)
    return self.head
head=LinkedList(arr)
print(head.head.next.next.value)
```
